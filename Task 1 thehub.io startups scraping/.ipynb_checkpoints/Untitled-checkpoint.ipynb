```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import openpyxl
import csv
from lxml import etree 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver import Keys, ActionChains
from selenium.webdriver.common.actions.action_builder import ActionBuilder
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException
import time
```

```python
def uncheck_denmark():
    label_locator = (By.XPATH, "//label[contains(., 'Denmark') and .//span[contains(., '(4365)')]]")
    WebDriverWait(driver, 40).until(EC.element_to_be_clickable(label_locator)).click()
def page_load():
    for j in range(1,635):
        print(j)
        try:
            elements = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'li.page-item a.page-link[href*="/startups?page="]')))
            if j < 3:
                elements[3].click()
            elif 3 <= j < 663:
                elements[4].click()
            print(driver.current_url)

        except StaleElementReferenceException:
            elements = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'li.page-item a.page-link[href*="/startups?page="]')))
            if j < 3:
                elements[3].click()
            elif 3 <= j < 663:
                elements[4].click()
            print(driver.current_url)
    driver.implicitly_wait(40)
```

```python
driver = webdriver.Chrome()
url = "https://thehub.io/startups"
url_personal = ""
driver.get(url)
uncheck_denmark()
time.sleep(5)
company_name=[]
location=[]
website=[]
founded=[]
employees=[]
industries=[]
business_model=[]
funding_state=[]

for i in range(634, 666):
    time.sleep(1)
    if i > 634:
        driver2 = webdriver.Chrome()
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')
        a_tags = soup.select('div.card.card-startup a')

        for a_tag in a_tags:
            href_value = a_tag.get('href')
            href_value = href_value[9:]
            url2 = url+href_value
            driver2.get(url2)
            #Company Name
            try:
                element = driver2.find_element(By.XPATH,"//h2[@class='startup-header__name']")
                WebDriverWait(driver2, 5).until(EC.visibility_of(element))
                company_name.append(element.text)
            except NoSuchElementException:
                driver2.refresh()
                element = driver2.find_element(By.XPATH,"//h2[@class='startup-header__name']")
                WebDriverWait(driver2, 5).until(EC.visibility_of(element))
                company_name.append(element.text)
            #location
            element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[1]/td[2]/b')
            WebDriverWait(driver2, 5).until(EC.visibility_of(element))
            location.append(element.text)
            #Website
            element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[2]/td[2]/a/b')
            WebDriverWait(driver2, 5).until(EC.visibility_of(element))
            website.append(element.text)
            #Founding Year
            element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[3]/td[2]/b')
            WebDriverWait(driver2, 5).until(EC.visibility_of(element))
            founded.append(element.text)
            #Number of Employees
            element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[4]/td[2]/b')
            WebDriverWait(driver2, 5).until(EC.visibility_of(element))
            employees.append(element.text)
            #Industries
            element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[5]/td[2]/b')
            WebDriverWait(driver2, 5).until(EC.visibility_of(element))
            industries.append(element.text)
            #Business Model
            try: 
                element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[6]/td[2]/b')
                WebDriverWait(driver2, 5).until(EC.visibility_of(element))
                business_model.append(element.text)
            except TimeoutException:
                pass
            #Funding State
            element = driver2.find_element(By.XPATH, '//*[@id="__layout"]/div/div[1]/div/div/content/div/div/div/section[1]/div/div/div/content/div/div[2]/table/tbody/tr[7]/td[2]/b')
            WebDriverWait(driver2, 5).until(EC.visibility_of(element))
            funding_state.append(element.text)            
        driver2.quit()
        elements = WebDriverWait(driver, 5).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'li.page-item a.page-link[href*="/startups?page="]')))
        if i<3:
            elements[3].click()
        elif i>=3 and i<663:
            elements[4].click()
        elif i>=663 and i<665:
            elements[5].click()
        else:
            print()   
    elif i<635:
        page_load()
    if i%5==0 and i>635:
        df = pd.DataFrame({'Column1': company_name,'Column2': location, 'Column3': website, 'Column4': founded, 'Column5': employees, 'Column6': industries, 'Column7': business_model, 'Column8': funding_state})
        # Save the DataFrame to an Excel file
        df.to_excel('firsthundred.xlsx', index=False)
df = pd.DataFrame({'Column1': company_name,'Column2': location, 'Column3': website, 'Column4': founded, 'Column5': employees, 'Column6': industries, 'Column7': business_model, 'Column8': funding_state})
# Save the DataFrame to an Excel file
df.to_excel('output.xlsx', index=False)
```
